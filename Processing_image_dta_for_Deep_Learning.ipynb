Resizing the image using Pillow Libary
  

```python
from PIL import Image #keyword for pillow libarry is PIL
```

```python
img = Image.open('/content/dog.jpg')
img_resized = img.resize((200,200))
#reducing the dimension of the org image
```

```python
img_resized.save('dog_image_resized.jpg')
#Saving the new image
```

```python
#displaying the image from numpy array
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_plot = plt.imshow(img_resized)
#plt is plotting mainly
```

```python
print(img_res.shape)
#Went to 200 by 200
```

Converting RgB images to Grayscale image using OpenCV

```python
import cv2
#importing OpenCV library
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
#RGB format to grayscale format
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
#the three is not there which means it shifted from 3 channel to 1 channel(RGB to Grayscale)
```
